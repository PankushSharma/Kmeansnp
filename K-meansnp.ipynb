```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.spatial.distance import cdist
X= (make_blobs(n_samples=1000,n_features=2,centers=3,cluster_std=1.8))[0]
centers=(make_blobs(n_samples=3,n_features=2,centers=3,cluster_std=1.8))[0]
X.shape
for i in range(10):
    euclidean_dist = cdist(X,centers,metric='euclidean')
    min_dist = np.argmin(euclidean_dist,axis=1)
    hot_encoding = pd.get_dummies(min_dist)
    hot_encoding1=np.array(hot_encoding)
    centers = (np.dot(X.T,hot_encoding1)/np.sum(hot_encoding1,axis=0)).T 
    print(centers.shape)
plt.scatter(X[:,0],X[:,1],c=min_dist)
plt.scatter(centers[:,0],centers[:,1],marker="+",s=500)
```
